```python
# visualization of addition of four sinusoidal waves, and the product demodulator
# finalized @20180417
# by HP

%matplotlib inline
import math
import ipywidgets as widgets
#from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact



@interact(amp_1=(0, 4, 1), omega_1=(0, 4, 1),  #phase_1=(-np.pi, np.pi, np.pi/4), 
         omega_2=(0, 4, 1), #phase_2=(-np.pi, np.pi, np.pi/4),
         fn = {'sin': np.sin, 'cos': np.cos,})

def wcomm_basic(omega_1=1, amp_1=1, #phase_1=0, 
         omega_2=1,  #phase_2=0, 
         fn=np.sin 
         ):
    

    amp_2=1
    phase_1=0
    phase_2=0
    
    # Create a new figure of size 8*6 point, using 100 dots per inch
    fig=plt.figure(figsize=(10,2), dpi=150)
    
    grid = plt.GridSpec(1,3)
    
    

    # 生成出图的原始数据
    X = np.linspace(0, 4 * np.pi, 512, endpoint=True)







    ax1 = fig.add_subplot(grid[0,0])
    # 设置X，Y 轴的范围
    ax1.set_xlim([0, 4 * np.pi])
    ax1.set_ylim([-2.5, 2.5])

    # 标记X，Y轴。注意:这里使用了Latex（to allow for nice rendering of the label）
    ax1.set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi, 
                2.5*np.pi, 3*np.pi, 3.5*np.pi, 4*np.pi])
    ax1.set_xticklabels(
               [r'$0$', r'$+\dfrac{1}{2}t$', r'$+t$', r'$+\dfrac{3}{2}t$', 
                r'$+2t$', r'$+\dfrac{5}{2}t$', r'$+3t$', r'$+\dfrac{7}{2}t$', r'$+4t$'])
    ax1.set_yticks([-2, -1, 0, +1, +2])
    ax1.set_yticklabels([r'$-2$', r'$-1$', r'$0$', r'$+1$', r'$+2$'])

    # 移动X，Y坐标轴(例如，坐标轴居中，显示四个象限)
    # 由于坐标轴包含四个(top/bottom/left/right), 将top和right坐标轴颜色设为none;
    ax1 = plt.gca() #gca() Get Current Axes; gcf() Get Current Figure.
    ax1.spines['right'].set_color('none')
    ax1.spines['top'].set_color('none')
    # 移动bottom和left至数据空间的coordinate 0
    ax1.xaxis.set_ticks_position('bottom')
    ax1.spines['bottom'].set_position(('data',0))
    ax1.yaxis.set_ticks_position('left')

    
    
    S_1= amp_1 * fn(omega_1 * X + phase_1)

    # cos的plot, 蓝色实线（线宽=2）&& sin的Plot, 绿色实线 （线宽=2）
    # 并添加图例
    
    ax1.plot(X, S_1, color="purple", linewidth=2.0, linestyle="-", label=r"Tx Signal: {i1}".format(i1=amp_1)+r"$s_{tx}$"+"({i2} t)".format(i2=omega_1))
    ax1.legend(loc='upper center',frameon=True)
    
    
    
    
    
    ax2 = fig.add_subplot(grid[0,1])
    ax2.set_xlim([0, 4 * np.pi])
    ax2.set_ylim([-2.5, 2.5])
    ax2.set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi, 
                2.5*np.pi, 3*np.pi, 3.5*np.pi, 4*np.pi])
    ax2.set_xticklabels(
               [r'$0$', r'$+\dfrac{1}{2}t$', r'$+t$', r'$+\dfrac{3}{2}t$', 
                r'$+2t$', r'$+\dfrac{5}{2}t$', r'$+3t$', r'$+\dfrac{7}{2}t$', r'$+4t$'])
    ax2.set_yticks([-2, -1, 0, +1, +2])
    ax2.set_yticklabels([r'$-2$', r'$-1$', r'$0$', r'$+1$', r'$+2$'])

    
    ax2 = plt.gca() 
    ax2.spines['right'].set_color('none')
    ax2.spines['top'].set_color('none')
    ax2.xaxis.set_ticks_position('bottom')
    ax2.spines['bottom'].set_position(('data',0))
    ax2.yaxis.set_ticks_position('left')
 
    S_2 = amp_2 * np.sin(omega_2 * X + phase_2)

    
    ax2.plot(X, S_2, color="blue", linewidth=2.0,linestyle="-" ,label=r"Rx Oscillator: $s_{rx}$"+"({i2} t)".format(i2=omega_2))
    ax2.legend(loc='upper center',frameon=True)
    
    
    
    
    ax3 = fig.add_subplot(grid[0,2])
    ax3.set_xlim([0, 4 * np.pi])
    ax3.set_ylim([-2.5, 2.5])
    ax3.set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi, 
                2.5*np.pi, 3*np.pi, 3.5*np.pi, 4*np.pi])
    ax3.set_xticklabels(
               [r'$0$', r'$+\dfrac{1}{2}t$', r'$+t$', r'$+\dfrac{3}{2}t$', 
                r'$+2t$', r'$+\dfrac{5}{2}t$', r'$+3t$', r'$+\dfrac{7}{2}t$', r'$+4t$'])
    ax3.set_yticks([-2, -1, 0, +1, +2])
    ax3.set_yticklabels([r'$-2$', r'$-1$', r'$0$', r'$+1$', r'$+2$'])

    
    ax3 = plt.gca() 
    ax3.spines['right'].set_color('none')
    ax3.spines['top'].set_color('none')
    ax3.xaxis.set_ticks_position('bottom')
    ax3.spines['bottom'].set_position(('data',0))
    ax3.yaxis.set_ticks_position('left')
 
    S_3 = S_1 * S_2

    ax3.plot(X, S_3, color="black", linewidth=2.0,linestyle="-" ,label=r"Product Modulator")
    ax3.legend(loc='upper center',frameon=True)
    





plt.show()
```
